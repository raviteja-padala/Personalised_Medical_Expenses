# Scraping Top medicines for each medical condition from medindia website and their prices from pharmacy website

```python
#importing libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import statistics
```

```python
#medical conditions
conditions  = ['commoncold','cough','fever','high-blood-pressure','diabetes','hypothyroidism','rheumatoid-arthritis']
```

```python
# appending medicines to list
med_list = []


for condition in conditions:

  url = 'https://www.medindia.net/drugs/medical-condition/'+condition+'.htm'
  response = requests.get(url)
  soup = bs(response.content, 'html.parser')

   #example code to find out all h1 tags

  for i in range(0,8):

    med= soup.find_all('h3')[i].text

    med_dict = {"Condition": condition , "Medicine ": med}
    med_list.append(med_dict)
```

```python
med_list[0]
```

```python
med_list[0]['Medicine ']
```

```python
# Dictionary to store medicines condition-wise
medicine_dict = {}

# Separating medicines into individual lists based on conditions
for item in med_list:
    condition = item['Condition']
    medicine = item['Medicine ']
    if condition in medicine_dict:
        medicine_dict[condition].append(medicine)
    else:
        medicine_dict[condition] = [medicine]

# Print the individual lists of medicines for each condition
for condition, medicines in medicine_dict.items():
    print(f"{condition}: {medicines}")
```

#Cold Medicine

```python
cold_links = []

for p in medicine_dict['commoncold']:
  website = 'https://www.apollopharmacy.in/search-medicines/'
  url = website + p
  cold_links.append(url)


names = []
Final_price = []


for f in range(len(cold_links)):
  response = requests.get(cold_links[f])
  soup = bs(response.content, 'html.parser')
  bigbox= soup.find_all("div", {"class":"ProductCard_pdHeader__ETKkp"})
  if bigbox == []:
    continue

  link='https://www.apollopharmacy.in'+bigbox[0].a['href']

  prodResult = requests.get(link)
  prodResult.encoding='utf-8'
  product_page = bs(prodResult.text, "html.parser")


  title= product_page.find_all("div", {"class":"PdpWeb_productDetails__am1oN"})[0].h1.text
  names.append(title)
  MRP = product_page.find_all("div", {"class":"MedicineInfoWeb_priceGroup__d8bys"})[0].h2.text
  discounted_price = product_page.find_all(class_="MedicineInfoWeb_medicinePrice__HPf1s")[0].contents[1]


    #print("Product_name=",title, "#PRICE=",price, sep = ' ')
    #print("\n")

  if len(MRP)>12:
        Final_price.append(discounted_price)
  else:
        Final_price.append(MRP)


# Stripping '₹' character from values in the list using a list comprehension
stripped_list = [value.replace('₹', '') for value in Final_price]
#converting to float
float_list = [float(value) for value in stripped_list]

cdata= {
    'Cold Medicines': names,
    'Price': stripped_list
}


float_list.sort()
print("Medicine Prices:",float_list)
print("mean:",statistics.mean(float_list))
print("median:", statistics.median(float_list) )

df = pd.DataFrame(cdata, index=range(1, len(names)+1))
df
```
